## Instalación en Windows

1. Instalar Chocolatey (si no está instalado):

Abre PowerShell como administrador y ejecuta:

```powershell
Set-ExecutionPolicy Bypass -Scope Process -Force; [System.Net.ServicePointManager]::SecurityProtocol = [System.Net.ServicePointManager]::SecurityProtocol -bor 3072; iex ((New-Object System.Net.WebClient).DownloadString('https://community.chocolatey.org/install.ps1'))
```

2. Instalar Git, Visual Studio Code y Python:

```powershell
choco install git
choco install vscode
choco install python3
```

3. Clonar el repositorio de apuntes:

```powershell
git clone https://github.com/mrey-profe/programacion-python
```

4. Entrar en la carpeta creada:

```powershell
cd programacion-python
```

5. Abrir el contenido con Visual Studio Code:

```powershell
code .
```

6. Instalar las extensiones recomendadas: Jupyter y Python

7. Dar permiso para instalar ipykernel

## Instalación en Linux

1. Actualizar los repositorios de Ubuntu:

```bash
sudo apt update
```

2. Instalar Python3, pip e ipykernel:

```bash
sudo apt install python3
sudo apt install python3-pip
sudo apt install python3-ipykernel
```

3. Instalar Visual Studio Code:

```bash
sudo snap install --classic code
```

4. Instalar Git:

```bash
sudo apt install git
```

5. Clonar el repositorio de apuntes:

```bash
git clone https://github.com/mrey-profe/programacion-python
```

6. Entrar en la carpeta creada:

```bash
cd programacion-python
```

7. Abrir el contenido con Visual Studio Code:

```bash
code .
```

8. Instalar las extensiones recomendadas: Jupyter y Python
