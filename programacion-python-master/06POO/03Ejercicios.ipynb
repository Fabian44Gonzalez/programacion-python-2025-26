# Ejercicio 6

## Clase Punto

Crea una clase `Punto` que modele un punto en un espacio bidimensional con las siguientes características:

1. Tendrá dos atributos privados, `x` e `y`, que guardan las coordenadas. Los valores de estos atributos serán enteros entre -100 y 100. Deben implementarse los métodos apropiados para acceder a estos atributos comprobando que los valores son correctos.

2. El constructor recibirá las coordenadas x e y. Si los datos pasados son incorrectos (o no se indican), creará un punto en (0,0).

3. Implementa los siguientes métodos:

   - `mover_horizontal(positivo: bool, valor: int) -> None`: Modificará la coordenada x de la siguiente manera:
     - Si `positivo` es `True`, incrementará su posición según `valor` (máximo 100)
     - Si `positivo` es `False`, decrementará su posición según `valor` (mínimo -100)

   - `mover_vertical(positivo: bool, valor: int) -> None`: Modificará la coordenada y de la siguiente manera:
     - Si `positivo` es `True`, incrementará su posición según `valor` (máximo 100)
     - Si `positivo` es `False`, decrementará su posición según `valor` (mínimo -100)

   - `obtener_cuadrante() -> int`: Devolverá un valor entre 1 y 4 (cada número representa a un cuadrante) según las siguientes condiciones:
     - Si x>0 y y>0: cuadrante 1
     - Si x<0 y y>0: cuadrante 2
     - Si x<0 y y<0: cuadrante 3
     - Si x>0 y y<0: cuadrante 4

   - Implementa el método `__str__` para que devuelva el punto como: "[valorX, valorY]"

## Clase Línea

Crea una clase `Línea` con las siguientes características:

1. Tendrá un atributo privado `coordenadas` que será una lista de dos objetos `Punto`.

2. El constructor recibirá dos puntos. Si los puntos no son válidos, se creará una línea con dos puntos en (0,0).

3. Implementa los siguientes métodos:

   - `mover_horizontal(positivo: bool, valor: int) -> bool`: Mueve los dos puntos horizontalmente.
     - Si `positivo` es `True`, mueve a la derecha
     - Si `positivo` es `False`, mueve a la izquierda
     - Devuelve `True` si se realizó el movimiento, `False` en caso contrario

   - `mover_vertical(positivo: bool, valor: int) -> bool`: Mueve los dos puntos verticalmente.
     - Si `positivo` es `True`, mueve hacia arriba
     - Si `positivo` es `False`, mueve hacia abajo
     - Devuelve `True` si se realizó el movimiento, `False` en caso contrario

   - Implementa el método `__str__` para que devuelva la línea como: "Punto A [valorX, valorY], Punto B [valorX, valorY]"

Nota: El movimiento no se realizará si `valor` no es un número positivo > 0 o si la línea se sale del espacio del plano (-100 a 100 en ambos ejes).
